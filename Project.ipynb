# 📊 Sales Data Analysis Project

## Objective
Analyze sales data to identify trends across products and cities.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 📁 Dataset

```python
# Create dataset
data = {
    "Date": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"],
    "Product": ["Laptop", "Mobile", "Tablet", "Laptop", "Mobile"],
    "Sales": [50000, 20000, 15000, 55000, 22000],
    "City": ["Hyderabad", "Chennai", "Bangalore", "Hyderabad", "Chennai"]
}

df = pd.DataFrame(data)
df
```

## 📦 Sales by Product

```python
product_sales = df.groupby("Product")["Sales"].sum()

product_sales.plot(kind="bar")
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

## 🌍 Sales by City

```python
city_sales = df.groupby("City")["Sales"].sum()

city_sales.plot(kind="bar")
plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```

## 📊 Visualization

```python
sns.barplot(x="Product", y="Sales", data=df)
plt.title("Product vs Sales")
plt.show()
```
